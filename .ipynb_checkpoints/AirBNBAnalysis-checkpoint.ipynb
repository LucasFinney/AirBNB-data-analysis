# Visualization 1: Distribution of Neighbourhoods

This is essentially just a point of curiosity for me, but I might as well check... since this is categorical, I think the best visualization will either be a pie-chart or a bar chart depending on how thin some of the slices get.

```python
nbhdCounts = listings['neighbourhood_cleansed'].value_counts()
nbhdCounts
```

Okay, now that is interesting. Maybe a Pareto-chart would be appropriate here?

```python
fig, ax1 = plt.subplots()

ax1.bar(nbhdCounts.index, nbhdCounts.values, color = 'blue')
ax1.set_ylabel('Frequency', color = 'blue')

plt.show
```

Okay, there are far too many neighbourhoods for this to be meaningful. Maybe just isolate the first ten or so, then have the remainder be "other"?

```python
# Select the top 10 most frequent
top10nbhd = nbhdCounts.head(10)

# Calculuate count for the rest
otherCount = nbhdCounts[10:].sum()

# Create new DataFrame
data = pd.concat([top10nbhd, pd.Series(otherCount, index = ['Other'])])

# Plot the modified DataFrame
plt.bar(data.index, data.values)
plt.xlabel('Rental Place')
plt.ylabel('Frequency')
plt.title('Distribution of Rental Places')

# Display the chart
plt.show()
```

Okay, still not what I'm looking for, though I can see that I will want to adjust my label display again, as before. Need to find the point where the "other" category is a similar height as the column next to it. 

Need n such that count(Xn) ~ sum(Xm, {m,n,length})

```python
# Calculate the cumulative sum of frequencies in descending order
cumulative_sum = nbhdCounts.sort_values(ascending=False).cumsum()

# Find the rank where the frequency is approximately equal to the sum of remaining frequencies
rank = cumulative_sum[cumulative_sum >= cumulative_sum.iloc[-1] / 2].idxmin()

print("Rank:", rank)

# Find the position or rank of the label in the sorted list
position = nbhdCounts.index.tolist().index(rank) + 1
print("Position:", position)
```

So, I want to take the top 15, then leave the rest as "other". Next thing I'll do is trim nbhd names for readability.

```python
#Take the top 15, separate out the rest
top15nbhd = nbhdCounts.head(15)

# Calculuate count for the rest
otherCount = nbhdCounts[15:].sum()

# Create new DataFrame
data = pd.concat([top15nbhd, pd.Series(otherCount, index = ['Other'])])



# Specify the desired number of characters to keep
desired_length = 10

# Trim the neighborhood names beyond the desired length
trimmed_labels = data.index.map(lambda x: x[:desired_length])

# Create a new DataFrame with trimmed labels and their corresponding counts
dataNew = pd.DataFrame({'Neighborhood': trimmed_labels, 'Frequency': data.values})

dataNew
```

Ugh, this isn't going to work quite well... I'll try a pie-chart visualization, but it looks like the big problem is that Waterfront is VASTLY more popular than other locations, and that there's a fairly gentle "slope" as we go downwards in frequency past that point.

```python
# Plot the pie chart
plt.pie(nbhdCounts, labels=nbhdCounts.index, autopct='%1.1f%%')

# Set the aspect ratio to make the pie chart circular
plt.axis('equal')

# Set a title for the pie chart
plt.title('Distribution of Neighborhoods')

# Display the chart
plt.show()
```
